### What are the bigrams in the following mini corpus? What are their MLEs?

```python
txt = \
"<s> How much wood would a wood chuck chuck if a wood chuck could chuck wood ? </s> <s> As much wood as a wood chuck could if a wood chuck could chuck wood . </s>"
```

```python
# List of bigrams in mini corpus
txt_words = txt.split()
list(nltk.bigrams(txt_words))
```

```python
bi_gram_frq = nltk.FreqDist(nltk.bigrams(txt_words))
```

```python
uni_gram_frq = nltk.FreqDist(txt_words)
```

```python
mle_bi_gram = dict()
for key,value in dict(bi_gram_frq).items():
    mle_bi_gram[key] = value / uni_gram_frq[key[0]]
```

```python
# MLE for each Bi-Gram
mle_bi_gram
```
